Ques 1. What is Simple Linear Regression?

Ans 1. Simple linear regression is a statistical method that models the relationship between one independent variable (X) and one dependent variable (Y) using a straight line. It's used to predict or estimate the value of the dependent variable (outcome) based on the value of the independent variable.

Key Concepts:

1. Independent Variable (X):
The predictor or explanatory variable. It's assumed to be the cause or driver of changes in the dependent variable.

2. Dependent Variable (Y):
The response or outcome variable. It's the variable being predicted or explained.

3. Linear Relationship:
The model assumes a straight-line relationship between the two variables, meaning that a change in the independent variable is associated with a proportional change in the dependent variable.

4. Line of Best Fit:
The regression line is drawn through the data points in a scatter plot, minimizing the overall distance between the data points and the line. This line represents the best linear approximation of the relationship.

5. Slope:
The slope (coefficient) of the line indicates the amount by which the dependent variable changes for each unit change in the independent variable.

6. Intercept:
The intercept represents the predicted value of the dependent variable when the independent variable is zero.

Ques 5. How do we calculate the slope m in Simple Linear Regression?

Ans 5. In simple linear regression, the slope m is calculated using the formula: m = r * (sy / sx), where r is the correlation coefficient, sy is the standard deviation of the dependent variable, and sx is the standard deviation of the independent variable.

Ques 7. How is the coefficient of determination (R) interpreted in Simple Linear Regression?

Ans 7. In simple linear regression, the coefficient of determination (R) indicates the proportion of the total variation in the dependent variable that is explained by the independent variable. It ranges from 0 to 1 (or 0% to 100%), with a higher R indicating a better fit of the regression model to the data.

1. R as a measure of goodness-of-fit:
R essentially tells you how well your regression line (the predicted values from your model) matches the actual data points. A higher R means your model is capturing more of the variation in the data.

2. Interpretation:
R = 0: The independent variable explains none of the variation in the dependent variable. The model's predictions are no better than simply using the mean of the dependent variable.

3. R = 1: The independent variable perfectly explains the variation in the dependent variable. The regression line passes through all the data points, and the model is a perfect fit.

4. R = 0.75: The independent variable explains 75% of the variation in the dependent variable. The model's predictions are significantly better than using the mean of the dependent variable.

Elaboration:

Ques 11. What is heteroscedasticity, and how does it affect the results of a Multiple Linear Regression model?

Ans 11. Heteroscedasticity in a multiple linear regression model refers to the non-constant variance of the error terms (residuals) across different values of the independent variables. This violates one of the core assumptions of ordinary least squares (OLS) regression, which assumes that the error terms have a constant variance (homoscedasticity). When heteroscedasticity is present, the OLS regression model can lead to biased and unreliable results.

hetero.jpeg


how heteroscedasticity affects the results of a multiple linear regression model:

Consequences of Heteroscedasticity:

1. Invalid Hypothesis Testing:
Heteroscedasticity can lead to incorrect p-values and unreliable t-tests for individual coefficients, as well as invalid F-tests for overall significance. The standard errors of the regression coefficients are underestimated, which can cause t-tests to falsely indicate statistical significance.

2. Inaccurate Standard Errors:
The OLS estimators are still unbiased in the presence of heteroscedasticity, but they are no longer the best linear unbiased estimators (BLUE). This means they are less efficient and their standard errors are biased.

3. Inefficient Estimates:
While the OLS estimators remain unbiased, they become inefficient due to the non-constant variance of the error terms. This means that the OLS estimators may not have the minimum variance among all linear unbiased estimators.

4. Unreliable Predictions:
Predictions made using a model with heteroscedasticity can be less reliable and accurate because the model's variance is not accurately estimated.

5. Distorted Statistical Inference:
The presence of heteroscedasticity can distort statistical inference, making it difficult to draw valid conclusions about the relationship between the independent and dependent variables.

Causes of Heteroscedasticity:

1. Omitted Variables: Failing to include relevant independent variables in the model can lead to heteroscedasticity.

2. Non-Linear Relationships: If the relationship between the independent and dependent variables is non-linear, the variance of the error terms may not be constant.

3. Outliers: Extreme values or outliers in the data can also cause heteroscedasticity.

4. Measurement Errors: Errors in measuring the variables can also introduce heteroscedasticity.


Ques 12. How can you improve a Multiple Linear Regression model with high multicollinearity?

Ans 12. To address multicollinearity in a multiple linear regression model, you can consider several strategies, including removing highly correlated variables, combining them, or using techniques like Principal Component Analysis (PCA) to reduce dimensionality and mitigate the issue according to Analytics Vidhya. Regularization techniques like Ridge regression can also be helpful.

The detailed explanation are as follows:

1. Identifying Multicollinearity:
Variance Inflation Factor (VIF): VIF measures how much the variance of a coefficient estimate is inflated due to multicollinearity. A VIF greater than 5 or 10 is often considered a problem, indicating high collinearity.

2. Addressing Multicollinearity:

a. Remove Highly Correlated Variables:
If you have variables that are highly correlated (VIF > 5 or 10), consider removing one of them from the model. You can use stepwise regression or best subsets regression to help with variable selection, or simply remove the variable with the highest VIF.

b. Combine Variables:
If you have two or more variables that are highly correlated and represent similar concepts, you can combine them into a single, more representative variable.

c. Principal Component Analysis (PCA):
PCA transforms your original correlated variables into a set of uncorrelated variables called principal components. This reduces the dimensionality of your data and can mitigate multicollinearity.

d. Regularization Techniques:
Ridge and Lasso regression are methods that penalize the coefficients of the independent variables. This can help to shrink the coefficients of highly correlated variables, reducing the impact of multicollinearity.

e. Increase Sample Size:
A larger sample size can help reduce the impact of multicollinearity because more data points lead to more variation among the predictors and more stable coefficient estimates.

f. Centering Variables:
Centering variables by subtracting their mean from each observation can sometimes reduce multicollinearity, especially when interaction terms are included in the model.

g. Consider Interaction Terms:
Adding interaction terms (the product of two independent variables) can sometimes reduce multicollinearity by capturing the combined effect of the variables.

h. Collect More Data:
Increasing the sample size can also help by providing more variation among the predictors and making the relationships between variables more stable.

Ques 13. What are some common techniques for transforming categorical variables for use in regression models?

Ans 13. Several techniques can transform categorical variables for use in regression models. Common methods include one-hot encoding, which creates binary variables for each category, and label encoding, which assigns a unique numerical value to each category. Additionally, techniques like dummy coding and ordinal encoding can be used, depending on the nature of the categorical data.

Various methods are as follows:

1. One-Hot Encoding:
This method creates a new binary variable (0 or 1) for each unique value within a categorical variable. For instance, if you have a "color" variable with values "red," "blue," and "green," one-hot encoding would create three new variables: "red," "blue," and "green," with each variable being 1 if the corresponding color is present and 0 otherwise.

2. Label Encoding:
This technique assigns a unique numerical label (e.g., 0, 1, 2) to each category. For example, "red" might be assigned 0, "blue" 1, and "green" 2.

3. Dummy Coding:
Similar to one-hot encoding, dummy coding creates binary variables. However, it drops one of the categories to avoid multicollinearity, according to a blog on Analytics Vidhya.
Ordinal Encoding:
If the categorical variable has a meaningful order (e.g., "low," "medium," "high"), ordinal encoding can assign numerical values to the categories in a sequence that reflects the order.

4. Target Encoding:
This approach replaces the original categorical variable with the mean or other statistics of the target variable for each category.

5. Feature Engineering:
This broader term encompasses creating new features from existing ones, including transforming categorical variables into numerical ones using various methods.

The choice of method depends on the nature of the categorical variable and the specific regression model being used. For example, if the categories are not ordered, one-hot encoding or dummy coding might be appropriate. If the categories have a natural order, ordinal encoding might be preferable.


Ques 14. What is the role of interaction terms in Multiple Linear Regression?

Ans 14. In Multiple Linear Regression, interaction terms are crucial for understanding how the effect of one independent variable on the dependent variable changes depending on the level of another independent variable. They allow the model to capture non-additive relationships between predictors.

Modifying the Relationship:
Interaction terms help determine if the relationship between a predictor variable and the outcome is the same across all levels of another predictor variable. For example, the effect of advertising spend might be different if it's combined with a marketing campaign.

Non-Additive Effects:
If the effect of one variable on the outcome depends on the level of another, it means the relationship is not simply additive. Interaction terms allow the model to account for this non-additive effect.

Interpreting Coefficients:
The coefficients of main effects (independent variables) and the interaction term are interpreted differently. The coefficient of the interaction term indicates how much the effect of one predictor changes for each unit change in the other predictor.

Examples:
Two-Way Interaction: A two-way interaction term captures the interaction between two variables (e.g., X1 * X2).

Three-Way Interaction: A three-way interaction term captures the interaction between three variables (e.g., X1 * X2 * X3).


Ques 15. How can the interpretation of intercept differ between Simple and Multiple Linear Regression?

Ans 15. In both simple and multiple linear regression, the intercept represents the predicted value of the dependent variable when all independent variables are zero. However, in simple linear regression, this interpretation is straightforward. In multiple linear regression, the intercept also represents the expected value of the dependent variable when all other predictors are held constant.

1. Simple Linear Regression:

a. Intercept:
The predicted value of the dependent variable (Y) when the independent variable (X) is zero. For example, if you're predicting sales based on advertising spend, the intercept represents the baseline sales when no advertising is done.

b. Single Slope:
The relationship between X and Y is represented by a single slope coefficient.

2. Multiple Linear Regression:

a. Intercept:
The predicted value of the dependent variable (Y) when all independent variables are zero. Importantly, this is the expected value of Y when all other independent variables in the model are also set to zero.

b. Multiple Slopes:
Each independent variable has its own slope coefficient, reflecting the unique impact of that variable on the dependent variable, holding all other variables constant.

c. Interpreting the Intercept in Multiple Regression:
The intercept is interpreted as the predicted value of Y when all independent variables are zero. In practical terms, this might be a baseline level of something (like sales) before any other factors are considered.

Key Differences and Considerations:

1. Predicting the Baseline:
In simple regression, the intercept is a baseline value for the independent variable. In multiple regression, it's the baseline for the combined effect of all variables.

2. Interpreting Slopes:
Multiple regression allows you to examine the independent effects of each variable on the dependent variable, while controlling for other variables.

3. Meaningful Intercept:
The intercept's practical meaning depends on whether the zero values for independent variables are realistic or within the range of your data.

In essence: While both intercepts represent the expected value of Y when all Xs are zero, the interpretation of the intercept's meaning changes in multiple regression as you consider the combined effect of multiple predictors, each with its own slope

Ques 16. What is the significance of the slope in regression analysis, and how does it affect predictions?

Ans 16. In regression analysis, the slope represents the change in the dependent variable for every one-unit change in the independent variable. 2. Direction of Relationship:
A positive slope indicates a positive correlation (as X increases, Y increases), while a negative slope indicates a negative correlation (as X increases, Y decreases).

3. Strength of Relationship:
A steeper slope suggests a stronger relationship between the variables, meaning a larger change in the dependent variable for a given change in the independent variable.

4. Predictive Power:
A significant slope (determined through hypothesis testing) indicates that the independent variable has a statistically significant influence on the dependent variable.

How the Slope Affects Predictions:

1. Accurate Predictions:
A slope that accurately reflects the true relationship between the variables allows the regression model to make more accurate predictions.

2. Outliers:
If the slope is inaccurate due to outliers or a poor model fit, predictions may be significantly off.

3. Adjusting for Changes:
The slope is crucial for understanding how the dependent variable will change when the independent variable is changed, enabling informed decisions or predictions based on variations in the independent variable.

4. Interpreting Predictions:
The slope helps interpret the meaning of predictions. For example, if the slope is 2.1, increasing the advertising expenditure by one dollar is predicted to increase revenue by 2.1 dollars.

Ques 18. What are the limitations of using R as a sole measure of model performance?

Ans 18. R alone is not a sufficient measure of model performance due to several limitations. It doesn't indicate model quality, reliability, or whether the chosen model is the best one. R can be misleading with overfitting, especially when adding irrelevant variables. It's also sensitive to non-linear relationships and might not capture predictive accuracy on new data.

limitations are as follows:

1. Overfitting:
R can increase simply by adding more predictors, even if those predictors don't provide meaningful explanatory power. This can lead to a model that fits the training data well but doesn't generalize well to new data, according to Number Analytics.

2. Doesn't guarantee predictive accuracy:
A high R doesn't mean the model will make accurate predictions on new data, says Analytics Vidhya.

3. Sensitivity to non-linear relationships:
R is primarily designed for linear models. It may not adequately capture the performance of models that include non-linear relationships.

4. Doesn't account for model complexity:
R doesn't penalize for adding more variables, so a model with many predictors might have a high R even if it's overly complex and doesn't generalize well.

5. Doesn't measure the direction of the relationship:
R tells you how well the model fits the data but not whether it's positive or negative.

6. Can be misleading with small datasets:
R might give misleading results when working with smaller datasets.

7. Not directly comparable across datasets:
R values are not directly comparable across different datasets or dependent variables.

Ques 19. How would you interpret a large standard error for a regression coefficient?

Ans 19. A large standard error for a regression coefficient suggests that the estimated coefficient is less precise and less reliable as an estimate of the true population value. It indicates that the coefficient could vary significantly across different samples of the data.

Elaboration:

1. Precision:
The standard error quantifies the variability of the coefficient estimate across different samples. A larger standard error means the estimate is likely to fluctuate more if you were to repeat the study with a different sample.

2. Reliability:
A large standard error implies less confidence in the precision of the coefficient estimate, meaning it may not be a reliable predictor of the true population effect.

3. Statistical Significance:
While a large standard error doesn't automatically mean the coefficient is not significant, it does make it more challenging to demonstrate statistical significance. A large standard error would likely lead to a higher p-value (or a lower t-statistic) for the coefficient, making it less likely to be considered significantly different from zero.

4. Multicollinearity:
In regression analysis, a large standard error can also be a sign of multicollinearity, where independent variables are highly correlated with each other. This makes it difficult to isolate the effect of one variable on the dependent variable, leading to inflated standard errors.

In essence, a large standard error indicates a greater degree of uncertainty around the coefficient estimate, suggesting it may not be a reliable or precise indicator of the true relationship between the independent and dependent variables.


Ques 20. How can heteroscedasticity be identified in residual plots, and why is it important to address it?

Ans 20. Heteroscedasticity in residual plots can be identified by observing a fan-shaped pattern where the spread of residuals (points on the plot) increases or decreases systematically as the fitted values (predicted values) change. This indicates that the variance of the errors is not constant across the range of predicted values. It's important to address heteroscedasticity because it violates the assumptions of many statistical tests, leading to inaccurate and unreliable inferences.

Elaboration:

1. Visual Inspection:
Plotting residuals against fitted values (predicted values) is a primary way to visually assess heteroscedasticity. If the residuals show a funnel or cone shape, it's a strong indicator of heteroscedasticity.

2. Statistical Tests:
Formal statistical tests like the Breusch-Pagan test and White test can be used to formally test for heteroscedasticity. These tests check if the variance of the residuals is related to the predictor variables in the model.

Addressing Heteroscedasticity:

1. Transforming Variables: Mathematical transformations can help stabilize the variance of the residuals.

2. Using Weighted Least Squares: This method gives more weight to observations with smaller variances, effectively reducing the influence of observations with larger variances.

3. Model Specification: Ensuring that the model includes all relevant variables and that the relationships are correctly specified can help reduce or eliminate heteroscedasticity.

4. Addressing Outliers: Outliers can sometimes contribute to heteroscedasticity, so carefully examining and addressing them may be necessary.

Ques 21. What does it mean if a Multiple Linear Regression model has a high R but low adjusted R?

Ans 21. A high R-squared and low adjusted R-squared in a multiple linear regression model suggest that the model might be overfitted, meaning it's capturing noise in the data rather than true relationships. While R-squared indicates a good fit with the added variables, the adjusted R-squared, which penalizes for including unnecessary predictors, shows that these added variables might not be providing significant improvements in the model's predictive power.

Elaboration:

1. R-squared:
Measures the proportion of variance in the dependent variable that's explained by the independent variables in the model. A high R-squared generally indicates a good fit, meaning the model captures a significant portion of the data's variability. However, R-squared can increase even when irrelevant variables are added, as it doesn't account for the complexity of the model.

2. Adjusted R-squared:
A modified version of R-squared that penalizes the addition of unnecessary predictors. It's calculated by considering the number of predictors and the sample size, ensuring that only variables that truly improve the model's fit are included.

3. Overfitting:
When a model is overfitted, it learns the training data too well, including noise and irrelevant patterns. This can lead to poor generalization to new, unseen data.

Implications of High R-squared and Low Adjusted R-squared:

1. The model's fit might be artificially inflated due to the inclusion of many variables.
2. Some of the added variables might not be statistically significant and may not contribute to the model's predictive accuracy.
3. The model may not generalize well to new data, as it's likely capturing noise specific to the training set.

Ques 22. Why is it important to scale variables in Multiple Linear Regression?

Ans 22. Scaling variables in multiple linear regression is beneficial for several reasons, including faster convergence of optimization algorithms like gradient descent, improved interpretability of coefficients, and preventing dominant features from overshadowing others, according to GeeksforGeeks and Medium.

Detailed explanation are as follows:

1. Faster Convergence:
When using iterative optimization algorithms like gradient descent to train the model, scaling features can significantly speed up the convergence process. If features have vastly different scales, the gradient descent might oscillate more and take longer to find the optimal solution.

2. Improved Coefficient Interpretability:
Scaling makes it easier to compare the relative importance of different features based on their coefficients. When features are on different scales, the magnitude of a coefficient can be misleading, and scaling ensures that coefficients reflect the true impact of each feature, according to Quora and Medium.

3. Preventing Feature Dominance:
Without scaling, features with larger ranges can have a disproportionate influence on the model. Scaling ensures that all features are treated equally, preventing dominant features from overshadowing less prominent ones.

4. Regularization:
Scaling is often necessary when using regularization techniques (like L1 or L2 regularization) because these techniques are sensitive to the scale of the features.

5. Homoscedasticity:
Scaling can help ensure that the error term (residuals) in the regression model has a constant variance (homoscedasticity), which is an assumption of multiple linear regression, according to Datatab.

6. General Good Practice:
While not always strictly necessary, scaling is generally considered good practice in machine learning, including multiple linear regression, due to its potential benefits for model performance and interpretability

Ques 24. How does polynomial regression differ from linear regression?

Ans 24. Polynomial regression differs from linear regression in the type of relationship it models between variables. Linear regression fits a straight line to the data, while polynomial regression fits a curve, allowing for more complex, non-linear relationships. This difference arises from how the independent variables are used in the model; polynomial regression uses powers and products of the independent variables to create a polynomial equation.

Elaboration:

1. Linear Regression:
Models a straight line relationship between the independent and dependent variables. The equation takes the form Y = a + bX, where Y is the dependent variable, X is the independent variable, a is the intercept, and b is the slope.

2. Polynomial Regression:
Models a curved relationship between the independent and dependent variables by including polynomial terms (e.g., X, X, etc.) in the equation. The equation takes the form Y = c0 + c1X + c2X + ... + cnX, where n is the degree of the polynomial and c is a set of coefficients.

3. Model Flexibility:
Polynomial regression is more flexible than linear regression because it can capture non-linear trends in the data. This can lead to better predictions when the relationship between variables is not linear.

4. Overfitting:
A potential drawback of polynomial regression is the risk of overfitting. If the degree of the polynomial is too high, the model might fit the training data too closely, including noise, and perform poorly on new, unseen data.

When to Use:
Use linear regression when the relationship between variables appears linear. Use polynomial regression when the relationship appears non-linear, and a curved line better fits the data.

Ques 28. What are the limitations of polynomial regression?

Ans 28. Polynomial regression, while powerful for modeling complex relationships, has limitations. One main issue is the risk of overfitting, especially with high-degree polynomials, where the model learns the training data too well and performs poorly on unseen data. Another limitation is the computational complexity and the challenge of selecting the optimal polynomial degree, which can impact model performance. Furthermore, polynomial regression can be more sensitive to outliers than linear regression.

Limitations are as follows:

1. Overfitting:
High-degree polynomials can fit the noise in the training data, leading to a model that is very specific to the training set and doesn't generalize well to new data.

2. Computational Complexity:
As the degree of the polynomial increases, the computational cost of training the model also increases.

3. Choosing the right degree:
Finding the optimal polynomial degree is crucial. A low-degree polynomial may underfit the data, while a high-degree polynomial risks overfitting, according to Applied AI Course. Techniques like cross-validation are often used to find the right balance.

4. Outlier sensitivity:
Polynomial regression can be more affected by outliers than linear regression, as the nonlinear nature of the model can amplify the impact of extreme data points.

5. Interpretability:
High-degree polynomials can be difficult to interpret, as the number of terms and their combinations can become complex.


Ques 29. What methods can be used to evaluate model fit when selecting the degree of a polynomial?

Ans 29. Several methods can be used to evaluate model fit when selecting the degree of a polynomial. Visual inspection, cross-validation, and information criteria like AIC and BIC are commonly used.

1. Visual Inspection:
Plot the data points with polynomials of varying degrees.
Visually assess how well the curves fit the data points.
A degree that captures the data's general trend without overfitting is preferred.

2. Cross-validation:
Split the data into training and validation (or testing) sets.
Train models with different polynomial degrees on the training set.
Evaluate the models on the validation set and choose the degree that yields the best performance.
This helps prevent overfitting by identifying the degree that generalizes well to unseen data.

3. Information Criteria (AIC and BIC):
These criteria balance model fit with complexity.
AIC (Akaike Information Criterion) and BIC (Bayesian Information Criterion) penalize models with more parameters (higher degree).
Lower AIC or BIC values suggest better models.
These can be used to compare models of different degrees.

4. Residual Analysis:
Examine the residuals (differences between predicted and actual values) for patterns or non-constant variance.
Ideally, residuals should be randomly distributed and have no obvious trends or outliers.
Residual analysis can help identify potential model misspecification or the need for more complex models.

5. Forward and Backward Selection:
Forward Selection: Start with a simple model (e.g., linear) and iteratively add polynomial terms until the model improvement is not significant.
Backward Selection: Start with a high-degree polynomial and iteratively remove terms until the model's performance degrades significantly.
These methods can help identify the optimal degree without overfitting.

6. Model Comparison:
Compare the performance of different models using metrics like R-squared, adjusted R-squared, or mean squared error.
R-squared measures the proportion of variance explained by the model, while adjusted R-squared accounts for the number of parameters in the model.
Lower mean squared error indicates better fit.

Example:
If plotting data with polynomials of increasing degrees shows that a quadratic (degree 2) model fits well and adding more terms (higher degrees) does not significantly improve the fit but leads to wild fluctuations, a quadratic model would likely be chosen.
Similarly, if AIC and BIC scores indicate that a lower-degree polynomial has the lowest value, it suggests that the lower-degree model is preferred.

Ques 31. How is polynomial regression implemented in Python?

Ans 31. Polynomial regression is implemented in Python using libraries such as scikit-learn. It extends linear regression by adding polynomial terms to the predictor variables. This allows the model to capture non-linear relationships between the independent and dependent variables.
Here's a breakdown of the implementation process:
1. Import Libraries: numpy for numerical operations, matplotlib.pyplot for plotting, sklearn.preprocessing for polynomial feature generation, sklearn.linear_model for linear regression, and sklearn.pipeline to combine steps.
2. Generate or Load Data:
Create or load a dataset where the relationship between variables is non-linear.
3. Create Polynomial Features:
Use PolynomialFeatures from sklearn.preprocessing to transform the input data by adding polynomial terms.
Specify the degree of the polynomial (e.g., degree=2 for a quadratic relationship).
4. Fit the Model:
Create a linear regression model using LinearRegression from sklearn.linear_model.
Fit the model to the transformed polynomial features and the target variable.
5. Make Predictions:
Use the fitted model to predict new values of the dependent variable based on new values of the independent variable.
6. Visualize Results:
Plot the original data points and the fitted polynomial regression curve to visualize the model's performance.

lets see this practically:

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.pipeline import make_pipeline

# Generate sample data
np.random.seed(0)
x = np.random.rand(100, 1) * 10
y = 2 + 3 * x + x**2 + np.random.randn(100, 1) * 10

# Create polynomial features
degree = 2
poly_features = PolynomialFeatures(degree)
x_poly = poly_features.fit_transform(x)

# Fit linear regression model
model = LinearRegression()
model.fit(x_poly, y)

# Predict
x_test = np.linspace(0, 10, 100).reshape(-1, 1)
x_test_poly = poly_features.transform(x_test)
y_pred = model.predict(x_test_poly)

# Plotting
plt.scatter(x, y, label='Data Points')
plt.plot(x_test, y_pred, color='red', label='Polynomial Regression')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.show()
```
